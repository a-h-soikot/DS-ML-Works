```python
not_allowed = ['n.a.', 'not available']

def convert_people(cell):
    if cell in not_allowed:
        return 'Sam Altman'
    return cell
    
def convert_eps(cell):
    if cell in not_allowed:
        return None
    return cell
    
df = pd.read_excel('stock_data.xlsx', 'Sheet1', converters = {
    'people' : convert_people, 'eps' : convert_eps
     })
df
```
