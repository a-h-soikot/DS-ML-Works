```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temparature')
plt.title('Weather')
plt.plot(x, y, color = 'g', linewidth=3.5, linestyle='dotted')
```

```python
plt.xlabel('Day')
plt.ylabel('Temparature')
plt.title('Weather')
plt.plot(x, y, color = 'r', marker='D', linestyle='', )
# blank line style is necessary if you don't want a line connecting dots
# alpha = visibility
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel("Day")
plt.ylabel('Temparature')
plt.title('Weather Report')

plt.plot(days, max_t, 'r', label = 'Max')
plt.plot(days, avg_t, 'orange', label = 'Avg')
plt.plot(days, min_t, 'b', label = 'Min')

plt.legend(loc='lower left')

plt.grid()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
plt.bar(company, revenue)

plt.ylabel('Revenue (B)')
plt.xlabel('Company Name')
plt.title('Tech Stock')
```

```python
profit=[40,2,34,12]

import numpy as np
xpos = np.arange(len(company))
xpos
```

```python
plt.ylabel('Revenue (B)')
plt.xlabel('Company Name')
plt.title('Tech Stock')

plt.xticks(xpos, company) # Set values(names) of companies in place of xpos
plt.bar(xpos-0.2, revenue, label="Revenue", width=0.4)
plt.bar(xpos+0.2, profit, label="Profit", width=0.4)
plt.legend()
```

```python
plt.ylabel('Companies')
plt.xlabel('Revenue / Profit (B)')
plt.title('Tech Stock')

xpos = np.flip(xpos)

plt.yticks(xpos, company) # Set values(names) of companies in place of xpos
plt.barh(xpos-0.2, revenue, label="Revenue", height=0.4)
plt.barh(xpos+0.2, profit, label="Profit", height=0.4, )
plt.legend()
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.8) # Histogram is frequency of occurence in a range
```

```python
plt.hist(blood_sugar, bins=[min(blood_sugar), 100, 125, max(blood_sugar)], rwidth=0.99, color='g')
```

```python
plt.hist(blood_sugar, bins=[min(blood_sugar), 100, 125, max(blood_sugar)], color='y', histtype='step')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist([blood_sugar_men, blood_sugar_women], bins=[40, 80, 120, 160], color=['g', 'y'],
          label=['Men', 'Women'], orientation='vertical')

plt.legend()
plt.grid()
```
