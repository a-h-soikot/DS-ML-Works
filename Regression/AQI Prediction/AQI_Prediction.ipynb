```python
import pandas as pd

df = pd.read_csv('air_quality_index_dataset.csv')
df
```

```python
df = df.drop(['RegionID'], axis='columns')
df.describe()
```

Removing Outlier

```python
columns = ['Temperature',	'Humidity',	'WindSpeed',	'PopulationDensity',	'IndustrialActivity',	'VehicleCount',	'GreenCoverage',	'Rainfall']
filtered_df = df.copy()

for column in columns:
  mean = df[column].mean()
  std = df[column].std()
  filtered_df = filtered_df[(filtered_df[column] > mean - 3 * std) & (filtered_df[column] < mean + 3 * std)]

filtered_df.describe()
```

```python
filtered_df
```

```python
df['ConstructionActivities'].unique()
```

```python
filtered_df['ConstructionActivities'] = filtered_df['ConstructionActivities'].map({'Low' : 0, 'Moderate' : 1, 'High' : 2 })
filtered_df
```

```python
X = filtered_df.drop(['AQI'], axis='columns')
Y = filtered_df['AQI']

from sklearn.preprocessing import RobustScaler, StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score

models = {
    'Linear Regression' : LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'Random Forest Regressor': RandomForestRegressor(random_state=10),
}


results = {}

for name, model in models.items():
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)

  results[name] = {
      'MSE' : mean_squared_error(y_test, y_pred),
      'R2 Score' : r2_score(y_test, y_pred)
  }
```

```python
from sklearn.neural_network import MLPRegressor

model = MLPRegressor(hidden_layer_sizes=(100, 50, 25, 5), activation='relu', solver='adam', max_iter=1000, random_state=42)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

results['MLP Regressor'] = {
  'MSE' : mean_squared_error(y_test, y_pred),
  'R2 Score' : r2_score(y_test, y_pred)
}

for name, result in results.items():
  print(f"{name}:")
  print(f"MSE: {result['MSE']}")
  print(f"r2_score: {result['R2 Score']}")
  print('-' * 30)
```
