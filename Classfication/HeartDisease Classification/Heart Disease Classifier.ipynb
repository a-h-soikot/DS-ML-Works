```python
# Excercise : https://github.com/codebasics/py/blob/master/ML/18_PCA/pca_exercise.md
import pandas as pd

df = pd.read_csv('heart.csv')
df
```

```python
print(df.isnull().sum())
df.dropna(inplace=True)
df.describe()
```

### Treat Outliers

```python
df_filtered = df.copy()

columns = ['RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak']

for column_name in columns:
    mean = df_filtered[column_name].mean()
    std = df_filtered[column_name].std()
    lower_bound = mean - 3 * std
    upper_bound = mean + 3 * std
    df_filtered = df_filtered[(lower_bound < df_filtered[column_name]) & (df_filtered[column_name] < upper_bound)]

df_filtered.describe()
```

```python
df_filtered['Sex'] = df_filtered['Sex'].map( { 'M' : 0, 'F' : 1 } )
df_filtered['ExerciseAngina'] = df_filtered['ExerciseAngina'].map( { 'N' : 0, 'Y' : 1 } )
df_filtered['ST_Slope'] = df_filtered['ST_Slope'].map( { 'Down' : 0, 'Flat' : 1, 'Up' : 2 } )
df_filtered
```

```python
df_filtered = pd.get_dummies(df_filtered, drop_first=True, dtype=int)
df_filtered
```

```python
X = df_filtered.drop('HeartDisease', axis='columns')
Y = df_filtered['HeartDisease']
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
x_scaled = scaler.fit_transform(X)
x_scaled
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, Y, test_size=0.2)
```

```python
x_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(x_train, y_train)
```

```python
model_rf.score(x_test, y_test)
```

### Now apply PCA and observe accuracy again

```python
from sklearn.decomposition import PCA

pca = PCA(0.9)
x_pca = pca.fit_transform(x_scaled)

x_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_pca, Y, test_size=0.2)

model_rf = RandomForestClassifier()
model_rf.fit(x_train, y_train)
```

```python
model_rf.score(x_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model_rf = LogisticRegression()
model_rf.fit(x_train, y_train)
model_rf.score(x_test, y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```
